```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 180, 100)

a = (-8.542)*10**(-4)
b = (1.146)*10**(-4)
c = 0
d = (8.830)*10**(-4)

def xy1(x):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(x))

def xy2r(x):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(x) + np.pi * 4/3 )

def xy2l(x):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(x) - np.pi * 4/3 )

def xy3r(x):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(x) - np.pi * 4/3 )

def xy3l(x):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(x) + np.pi * 4/3 )


def zx1(x):
    return (c+a)/2+(c-a)/2*np.cos(2*np.radians(x))

def zx2(x):
    return (4*c+a+3*b)/8+(4*c-(a+3*b))/8*np.cos(2*np.radians(x))+3**(1/2)/2*d*np.sin(2*np.radians(x))

def zx3(x):
    return (4*c+a+3*b)/8+(4*c-(a+3*b))/8*np.cos(2*np.radians(x))-3**(1/2)/2*d*np.sin(2*np.radians(x))


def yz1r(x):
    return (c+b)/2+(c-b)/2*np.cos(2*np.radians(x)) + d*np.sin(2*np.radians(x))

def yz1l(x):
    return (c+b)/2+(c-b)/2*np.cos(2*np.radians(x)) - d*np.sin(2*np.radians(x))

def yz23r(x):
    return (4*c+3*a+b)/8+(4*c-(3*a+b))/8*np.cos(2*np.radians(x))-d/2*np.sin(2*np.radians(x))

def yz23l(x):
    return (4*c+3*a+b)/8+(4*c-(3*a+b))/8*np.cos(2*np.radians(x))+d/2*np.sin(2*np.radians(x))

#figure()でグラフを表示する領域をつくり，figというオブジェクトにする．
fig = plt.figure()

#add_subplot()でグラフを描画する領域を追加する．引数は行，列，場所
ax1 = fig.add_subplot(4, 2, 1)
ax2 = fig.add_subplot(4, 2, 2)
ax3 = fig.add_subplot(4, 2, 3)
ax4 = fig.add_subplot(4, 2, 4)
ax5 = fig.add_subplot(4, 2, 5)
ax6 = fig.add_subplot(4, 2, 6)
ax7 = fig.add_subplot(4, 2, 7)
ax8 = fig.add_subplot(4, 2, 8)

ax1.plot(x, xy1(x)*10**4, color='yellow')
ax1.set_xlabel(r"$Rotation \quad angle$") 
ax1.set_ylabel(r"$K(\times 10^4)$") 
ax1.set_title('xy1')

ax2.plot(x, xy2r(x)*10**4, color='red')
ax2.plot(x, xy2l(x)*10**4, color='blue')
ax2.set_xlabel(r"$Rotation \quad angle$") 
ax2.set_ylabel(r"$K(\times 10^4)$") 
ax2.set_title('xy2r,xy2l')

ax3.plot(x, xy3r(x)*10**4, color='red', label=xy3r)
ax3.plot(x, xy3l(x)*10**4, color='blue', label=xy3l)
ax3.set_xlabel(r"$Rotation \quad angle$") 
ax3.set_ylabel(r"$K(\times 10^4)$") 
ax3.set_title('xy3r,xy3l')

ax4.plot(x, zx1(x)*10**4, color='yellow', label=zx1)
ax4.set_xlabel(r"$Rotation \quad angle$") 
ax4.set_ylabel(r"$K(\times 10^4)$") 
ax4.set_title('zx1')

ax5.plot(x, zx2(x)*10**4, color='yellow', label=zx2)
ax5.set_xlabel(r"$Rotation \quad angle$") 
ax5.set_ylabel(r"$K(\times 10^4)$") 
ax5.set_title('zx2')

ax6.plot(x, zx3(x)*10**4, color='yellow', label=zx3)
ax6.set_xlabel(r"$Rotation \quad angle$") 
ax6.set_ylabel(r"$K(\times 10^4)$") 
ax6.set_title('zx3')

ax7.plot(x, yz1r(x)*10**4, color='red', label=yz1r)
ax7.plot(x, yz1l(x)*10**4, color='blue', label=yz1l)
ax7.set_xlabel(r"$Rotation \quad angle$") 
ax7.set_ylabel(r"$K(\times 10^4)$") 
ax7.set_title('yz1r,yz1l')

ax8.plot(x, yz23r(x)*10**4, color='red', label=yz23r)
ax8.plot(x, yz23l(x)*10**4, color='blue', label=yz23l)
ax8.set_xlabel(r"$Rotation \quad angle$") 
ax8.set_ylabel(r"$K(\times 10^4)$") 
ax8.set_title('yz23r,yz23l')

fig.tight_layout()              #レイアウトの設定
plt.show()
```
