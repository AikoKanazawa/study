```python
fig, axes = plt.subplots(1, 1, figsize=(5, 5), constrained_layout=True)
axes.set_title(r"$The angle dependencies of peak$")
axes.set_xlabel(r"$Rotation angle$", fontsize=15) 
axes.set_ylabel(r"$Knight Shift$", fontsize=15)

axes.set_xlim([-5, 95])
axes.set_ylim([-0.5, 0])
axes.set_xticks(np.linspace(-5, 95, 11))
axes.set_yticks(np.linspace(-0.5, 0, 11))
axes.grid()

x = np.linspace(-5, 95, 100)

a = -0.11
b = -0.11
c = -0.25
d = -0.01

def y1(x):
    return (a+b)/2 + (a-b)*np.cos(2*np.radians(x))/2

def y2(x):
    return (b+c)/2 + (b-c)*np.cos(2*np.radians(x))/2 + d*np.sin(2*np.radians(x))

def y3(x):
    return (c+a)/2 + (c-a)*np.cos(2*np.radians(x))/2

ims=[]
im1 = axes.plot(x, y1(x), color='orange')
im2 = axes.plot(x, y2(x), color = 'magenta')
im3 = axes.plot(x, y3(x), color='blue')
axes.legend(['$ab$','$bc$', '$ca$'])

plt.show()
```
