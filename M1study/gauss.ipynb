```python
def gauss(x,a,mu,sigma):
    return a*np.exp(-(x-mu)**2 / (2*sigma**2))

fig = plt.figure()

A = fig.add_subplot(111)
A.grid(color="k",linestyle="dotted")
A.set_title("gauss function", fontsize = 16)
A.set_xlabel("x", fontsize = 14)
A.set_ylabel("y", fontsize = 14)

A.set_xlim([-8, 8])
A.set_ylim([0, 1.2])

x = np.arange(-8, 8, 0.1)

y1 = gauss(x,1,0,1)
y2 = gauss(x,0.5,2,2)
y3 = gauss(x,0.5,-2,2)

A.plot(x, y1, color="deeppink", label="a=1.0, =0, =1")
A.plot(x, y2, color="lime", label="a=0.5, =2, =2")
A.plot(x, y3, color="aqua", label="a=0.25, =2, =2")

A.legend()
plt.show()
```

```python
def N(x,mu,sigma):
    return (1/np.sqrt(2*np.pi*sigma))*np.exp(-(x-mu)**2 / (2*sigma**2))

fig = plt.figure()

A = fig.add_subplot(111)
A.grid(color="k",linestyle="dotted")
A.set_title("normal distribution", fontsize = 16)
A.set_xlabel("x", fontsize = 14)
A.set_ylabel("y", fontsize = 14)

A.set_xlim([-4, 4])
A.set_ylim([0, 1])

x = np.arange(-4, 4, 0.01)

y1 = N(x,0,0.2)
y2 = N(x,0,1)
y3 = N(x,-2,0.5)
y4 = N(x,2,0.5)

A.plot(x, y1, color="red", label="=0, =0.2")
A.plot(x, y2, color="blue", label="=0, =1")
A.plot(x, y3, color="green", label=" =-2, =0.5")
A.plot(x, y4, color="orange", label=" =2, =0.5")

A.legend()
plt.show()
```

```python
a = (-8.542)*10**(-4)
b = (1.146)*10**(-4)
c = 0
d = (8.830)*10**(-4)

def xy1(t):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(t))

def xy2r(t):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(t) + np.pi * 4/3 )

def xy2l(t):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(t) - np.pi * 4/3 )

def xy3r(t):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(t) - np.pi * 4/3 )

def xy3l(t):
    return (a+b)/2+(a-b)/2*np.cos(2*np.radians(t) + np.pi * 4/3 )


def zx1(t):
    return (c+a)/2+(c-a)/2*np.cos(2*np.radians(t))

def zx2(t):
    return (4*c+a+3*b)/8+(4*c-(a+3*b))/8*np.cos(2*np.radians(t))+3**(1/2)/2*d*np.sin(2*np.radians(t))

def zx3(t):
    return (4*c+a+3*b)/8+(4*c-(a+3*b))/8*np.cos(2*np.radians(t))-3**(1/2)/2*d*np.sin(2*np.radians(t))


def yz1r(t):
    return (c+b)/2+(c-b)/2*np.cos(2*np.radians(t)) + d*np.sin(2*np.radians(t))

def yz1l(t):
    return (c+b)/2+(c-b)/2*np.cos(2*np.radians(t)) - d*np.sin(2*np.radians(t))

def yz23r(t):
    return (4*c+3*a+b)/8+(4*c-(3*a+b))/8*np.cos(2*np.radians(t))-d/2*np.sin(2*np.radians(t))

def yz23l(t):
    return (4*c+3*a+b)/8+(4*c-(3*a+b))/8*np.cos(2*np.radians(t))+d/2*np.sin(2*np.radians(t))

o = float(13.433)*float(0.7430)
```

```python
t = 45

XY1 = o*(1+(xy1(t)))
XY2R = o*(1+(xy2r(t)))
XY2L = o*(1+(xy2l(t)))
XY3R = o*(1+(xy3r(t)))
XY3L = o*(1+(xy3l(t)))
ZX1 = o*(1+(zx1(t)))
ZX2 = o*(1+(zx2(t)))
ZX3 = o*(1+(zx3(t)))
YZ1R = o*(1+(yz1r(t)))
YZ1L = o*(1+(yz1l(t)))
YZ23R = o*(1+(yz23r(t)))
YZ23L = o*(1+(yz23l(t)))

print(XY1)
print(XY2R)
print(XY2L)
print(XY3R)
print(XY3L)
```
