```python
x = np.linspace(-180, 180, 100)

a = -8.542*10**(-4)
b = 1.146*10**(-4)
c = 0
d = 8.830*10**(-4)

def y1(x):
    return (a+b)/2 + (a-b)*np.cos(2*np.radians(x))/2

def y2(x):
    return (b+c)/2 + (b-c)*np.cos(2*np.radians(x))/2 + d*np.sin(2*np.radians(x))

def y3(x):
    return (b+c)/2 + (b-c)*np.cos(2*np.radians(x))/2 - d*np.sin(2*np.radians(x))

def y4(x):
    return (c+a)/2 + (c-a)*np.cos(2*np.radians(x))/2


o = float(13.433)*float(0.7430)
K1 = o*(1+(y1(x)))
K2 = o*(1+(y2(x)))
K3 = o*(1+(y3(x)))
K4 = o*(1+(y4(x)))

#figure()でグラフを表示する領域をつくり，figというオブジェクトにする．
fig = plt.figure()

#add_subplot()でグラフを描画する領域を追加する．引数は行，列，場所
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(x, K1, color='yellow')
ax1.axhline(o, color='magenta')
ax1.set_title('ab')
ax2.plot(x, K2, color='red')
ax2.axhline(o, color='magenta')
ax2.set_title('bc+')
ax3.plot(x, K3, color='blue')
ax3.axhline(o, color='magenta')
ax3.set_title('bc-')
ax4.plot(x, K4, color='yellow')
ax4.axhline(o, color='magenta')
ax4.set_title('ca')

fig.tight_layout()              #レイアウトの設定
plt.show()
```
