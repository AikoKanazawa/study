```python
from math import pi, sin, cos
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib nbagg
```

```python
def t_dep(i, x, y, line1, line2):
    line1.set_data(x[i], y[i])
    line2.set_data(x[:i], y[:i])
```

```python
thd = 45
th = thd * (pi/180)
n = 30
t = np.linspace(0, 2*sin(th), n)
x = t*cos(th)
y = t*sin(th)-0.5*t**2
fig = plt.figure()
plt.xlim(0,1)
plt.ylim(0,0.5)
line1, = plt.plot(x[0], y[0], 'ro', ms=10)
line2, = plt.plot(x[0], y[0], 'b--', lw=1)
ani = animation.FuncAnimation(fig, t_dep, fargs=(x, y, line1,line2),interval=100, frames=n, repeat=False)

from IPython.display import HTML
HTML(ani.to_jshtml())
```
