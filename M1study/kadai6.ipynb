```python
fig, axes = plt.subplots(1, 1, figsize=(5, 5), constrained_layout=True)
axes.set_title(r"Complex compliance($\lambda=1$)")
axes.set_xlabel(r"$log_{10}(\omega)$", fontsize=15) 
axes.set_ylabel(r"$GJ^{(1)},GJ^{(2)}$", fontsize=15)

axes.set_xlim([0.01, 1])
axes.set_ylim([0, 1])
axes.set_xticks(np.linspace(0.01, 1, 11))
axes.set_yticks(np.linspace(0, 1, 21))
axes.grid()

x = np.linspace(0.01, 1, 100)

l = 10

def y1(x):
    return 1/(1+x*x*l*l)

def y2(x):
    return x*l/(1+x*x*l*l)

ims=[]
axes.set_xscale('log')
im1 = axes.plot(x, y1(x), color='orange')
im2 = axes.plot(x, y2(x), color='blue')
axes.legend(['$GJ^{(1)}$','$GJ^{(2)}$'])

plt.show()
```

```python
N = 40
xmin = -2
xmax = 2
def H2(x):
   if -1 < x and x < 1: return 0
   else: return 1
p = np.linspace( xmin, xmax, N)
plt.plot( p, [H2(p[k]) for k in range(N)] )
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(5, 5), constrained_layout=True)
axes.set_title(r"$Carrier density$")
axes.set_xlabel(r"$x \mathrm{cm}$", fontsize=15) 
axes.set_ylabel(r"Carrier density", fontsize=15)

axes.set_xlim([-2.2*10**(-5), 2.2*10**(-5)])
axes.set_ylim([0, 1.5*(10**17)])
axes.set_xticks(np.linspace(-2.2*10**(-5), 2.2*10**(-5), 11))
axes.set_yticks(np.linspace(0, 1.5*(10**17), 16))
axes.grid()

xp = -2.360*(10**-6)
x0 = 0.0
xn = 1.574*(10**-6)
T = 38.67
Vbi = 0.8399
pn = 7.848 *(10**2)
np = 1.177 *(10**3)
nn = 1.5*(10**17)
pp = 10.0**17

# x1 = np.linspace(-2.2*10**(-5), xp, 100)
# x2 = np.linspace(xp, x0, 100)
# x3 = np.linspace(x0, xn, 100)
# x4 = np.linspace(xn, 2.2*10**(-5), 100)

x1 = [-2.2*10**(-5), -2.0*10**(-5), -1.9*10**(-5), xp]
x2 = [xp, -2.0*10**(-6), -1.9*10**(-6), 0]
x3 = [0, 1.0*10**(-6), xn]
x4 = [xn, 2.0*10**(-5), 2.2*10**(-5)]

y1 = pp

def y2(x):
    return y1 * np.exp(-0.6 * T * Vbi * ((x - xp) / (x0 - xp))**2)

def y3(x):
    return pn * np.exp(0.6 * T * Vbi * ((xn - x) / (xn - x0))**2)

y4 = pn

ims=[]
im1 = axes.axhline(y1, color='orange')
im2 = axes.plot(x2, y2(x2), color='blue')
im3 = axes.plot(x3, y3(x3), color='blue')
im4 = axes.axhline(y4, color='orange')
# axes.legend(['$GJ^{(1)}$','$GJ^{(2)}$'])

plt.show()
```
