```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.optimize import curve_fit
```

```python
# CSVファイルの読み込み
csv_file_path = 'xy(1).csv'
data = pd.read_csv(csv_file_path)

# xとyのデータを抽出
x_data = data.iloc[:, 0].values  # CSVファイルの1列目をxとして使用
y_data = data.iloc[:, 1].values  # CSVファイルの2列目をyとして使用
```

```python
# データのプロット
plt.scatter(x_data, y_data, label='Data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Visualization')
plt.legend()
plt.show()
```

```python
# 線形モデルの定義
def linear_model(x, a, b):
    return a * x + b

# 初期値の推定
# ここでは単純にデータの範囲を基に初期値を設定
a_initial = (y_data[-1] - y_data[0]) / (x_data[-1] - x_data[0])
b_initial = y_data[0] - a_initial * x_data[0]
initial_params = [a_initial, b_initial]

print("Estimated Initial Parameters: a =", a_initial, ", b =", b_initial)

# フィッティング
params, params_covariance = curve_fit(linear_model, x_data, y_data, p0=initial_params)

# フィッティング結果のプロット
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, linear_model(x_data, *params), label='Fitted function', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data and Fitted Linear Function')
plt.legend()
plt.show()

# フィッティングパラメータの表示
print("Fitted Parameters: a =", params[0], ", b =", params[1])
```
