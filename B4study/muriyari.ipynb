```python
### 複数のグラフを重ねて表示
import numpy as np
import matplotlib.pyplot as plt
import math

def y1(amp, x):
    return amp*np.exp(-x+4)

def y2(amp, x):
    return amp*np.exp(-x)

def y3(amp, x):
    return amp*np.exp(-6*x/2)*(1/(4*math.sqrt(2)))*(2-6*x)*math.pow(6, 3/2)

amp = math.pow(math.pi, -1/2)

x = np.linspace(4, 9, 1000)
y_1 = -4*(math.pi)*(x-4)*(x-4)*y1(amp,x)*y1(amp,x)

x = np.linspace(0, 9, 1000)
y_2 = 4*(math.pi)*x*x*y2(amp,x)*y2(amp,x)

c1, c2 = 'red', 'green'
l1, l2 = '"$\phi_{1s}$', '"$\phi_{1s}$'

xl = 'x'
yl = 'y'

#グラフを表示する領域を，figオブジェクトとして作成.
fig = plt.figure(figsize = (6,4), facecolor='lightblue')

x = np.linspace(0, 9, 1000)
plt.plot(x+4, y_1, color=c1, label=l1)
plt.plot(x, y_2, color=c2, label=l2)
plt.plot(x, y_1+y_2)

#ラベルを追加
# plt.xlabel(xl1)
# plt.ylabel(yl1)

# 凡例表示
plt.legend(loc = 'upper right')

plt.show()

```
