```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = int(input("Enter number of runs:"))
n = int(input("Enter number of steps:"))

trj = np.zeros((m, n))

for i in range(m):
    x = 0.0
    for j in range(n):
        x += (random.random() - 0.5) * 2
        trj[i,j] = x

for i in range(20):
    plt.plot(trj[i,:])
plt.show()

plt.hist(trj[:,-1], bins=50)
#plt.savefig("rw01hist.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# 1
N = 100
walkers = []  # HERE
for j in range(N):
    def randomwalk1D(n):
        x, t = 0, 0
        # Generate the time points [1, 2, 3, ... , n]
        time = np.arange(n + 1)
        position = [x]
        directions = [1, -1]
        for i in range(n):
            # Randomly select either +1 or -1
            step = np.random.choice(directions)
            
            # Move the object up or down
            if step == 1:
                x += 1
            elif step == -1:
                x -= 1
            # Keep track of the positions
            position.append(x)
        return time, position
    rw = randomwalk1D(1000)
    walkers.append(rw)
    plt.plot(rw[0], rw[1], 'r-', label="rw")
walkers = np.array(walkers)  # HERE
plt.plot(walkers[0][0], walkers[:, 1].mean(axis=0), 'b-', label="mean")  # HERE
plt.show()
```

```python
m = int(input("Enter number of runs:"))
n = int(input("Enter number of steps:"))

nplot = 20
pintv = m / nplot
xsq = np.zeros(n)

for i in range(m):
    x = 0.0
    xs = [x]
    for j in range(n):
        x += (random.random() - 0.5) * 2
        xsq[j] += x**2
        xs.append(x)
    if i % pintv == 0: plt.plot(xs) 

plt.show()

plt.plot(xsq/m)
#plt.savefig("rw01msd.png")
plt.show()
```
